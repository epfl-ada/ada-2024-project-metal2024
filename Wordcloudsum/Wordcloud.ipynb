### WW2

```python

ww2_keywords = {
    "allies": 3, "allied": 3, "auschwitz": 5
}

synonyms = {
    "second world war": "world war ii"
}

def classify_summary(summary, keyword_dict, threshold=5):
    summary_lower = summary.lower()
    # Replace synonyms if any
    for syn, replacement in synonyms.items():
        summary_lower = summary_lower.replace(syn, replacement)
        
    score = 0
    for kw, weight in keyword_dict.items():
        count = summary_lower.count(kw)
        if count > 0:
            score += weight * count
    return "WW2" if score >= threshold else "Not WW2"

# Load the full dataset
df = pd.read_csv("/Users/mattice/Desktop/DonDada/ada-2024-project-metal2024/data/processed_plot_summaries.csv")

# Take the first 35 summaries
df_subset = df

# Apply classification to the subset
df_subset["Classification"] = df_subset["Summary"].apply(lambda x: classify_summary(x, ww2_keywords, threshold=5))

# If you want to inspect or save the results:
print(df_subset[["Wikipedia movie ID", "Summary", "Classification"]])

# (Optional) save the subset results to a new file
df_subset.to_csv("processed_movies_classWW2.csv", index=False)
```

```python
count_ww2 = (df_subset["Classification"] == "WW2").sum()
print("Number of WW2 movies:", count_ww2)
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import nltk
from nltk.corpus import stopwords
import re

# Download NLTK stopwords if you haven't yet
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Assuming df_subset is already created and has the Classification column
ww2_df = df_subset[df_subset["Classification"] == "WW2"]

# Combine all WW2 summaries into a single string
all_ww2_text = " ".join(ww2_df["Summary"].tolist())

# Optional: Clean and preprocess the text
def clean_text(text):
    # Remove non-alphanumeric characters and lower text
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = text.lower()
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    # Remove stopwords
    words = [w for w in text.split() if w not in stop_words]
    return ' '.join(words)

all_ww2_text_clean = clean_text(all_ww2_text)

### 1. Create a Word Cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_ww2_text_clean)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bicubic')
plt.axis("off")
plt.title("Word Cloud of WW2 Summaries")
plt.show()

### 2. TF-IDF Transformation
vectorizer = TfidfVectorizer(stop_words='english', max_features=500)
X_tfidf = vectorizer.fit_transform(ww2_df["Summary"].apply(clean_text))

### 3. Perform LDA for Topic Modeling
# Specify number of topics
n_topics = 5
lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
lda.fit(X_tfidf)

# Print out top words per topic
feature_names = vectorizer.get_feature_names_out()
for topic_idx, component in enumerate(lda.components_):
    # Sort words by importance in this topic
    top_keywords = [feature_names[i] for i in component.argsort()[:-11:-1]]
    print(f"Topic {topic_idx+1}: {', '.join(top_keywords)}")
```
